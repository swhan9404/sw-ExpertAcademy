```python
리터럴 : 소스코드 상에서 내장 자료형의 상수 값을 나타내는 용어

정수형의 길이는 무제한, 메모리의 사용이 가능한 곳까지 쓸 수 있음
0o : 8진수, 0x : 16진수, 0b : 2진수

허수형은 j접미사를 사용
허수형 리터럴 내의 _는 무시됨

문자열 포맷팅 : 문자열 내 사용된 문자열 표시 유형(문자열 포맷코드)를 특정한 값으로 변경하는 기법
%-포맷팅
%s : 문자열 포맷
%c : 문자포맷, 정수를 유니코드 문자로 변환해 출력
%d : 10진 정수 출력
%o : 8진수로 출력
%x : 16진수로 출력
%f : 부동소수점 숫자로 출력, 소수점 이하 6자리 정밀도를 기본값으로 가짐
%% : % 문자 자체를 출력 

방식1.
>>> "이름 : %s: % "홍길동"
방식2. 튜플 사용
>>> "이름: %s \n나이 : %s 세" % ("홍길동", 20)
방식3. 딕셔너리 사용
>> "이름 : %(name)s \n나이 : %(age)s 세" %{"name":"홍길동", "age":20}
방식4. 우측정렬
"%10s" % "우측정렬" #'      우측정렬'
방식5. 좌측정렬
"%-10s" % "좌측정렬" #'좌측정렬      ' 문자열 폭이 10, 정려 방향이 좌측임
방식6. 부동소수점 출력 폭지정, 소수점 이하의 정밀도 표현적용
"%10.2f" %3.1415192 #'      3.14' 
"%010.2f" %3.141592 #'0000003.14' 부동소수점을 소수점을 포함한 전체 자리수는 10, 소수점 이하 두자리까지 표시하고, 앞의 남은 자리를 0으로 채움

str.format() 
방식 1. 값의 위치를 이용
"이름 {0}, 나이{1}세".format("홍길동", 20)
"{0:c} => {1}.format(97,97) # 문자열 내의 {} 에서 : 뒤에 문자열 표시 유형을 사용해 포맷지정
방식2. 이름 이용
"이름 : {name}, 나이 : {age}세".format(name="홍길동", age=20)
방식3. 문자열 폭, 정렬방향 지정
"{0:<10}".format("좌측정렬") #'좌측정렬      '
"{0:>10}".format("우측정렬") #'      우측정렬'
"{0:^10}".format("중앙정렬") #'   중앙정렬   '
"{0:*^10}".format("중앙정렬" #'***중앙정렬***'   #*은 공백을 채울 문자를 의미
방식5 : 부동소수점 숫자 폭, 소수점 이하 정밀도 표현
"{0:0.2f}.format(3.1415192) #'3.14'
"{0:10.2f}".format(3.1415192) #'      3.14'
"{0:010.2f}.".format(3.1415192) #'0000003.14'
"{{ {0:.1f} }}".format(98.5) #{ 98.5 }'
```

```python
4. 변수

변수의 제거
Garbage Collector에 의해 자동으로 관리
> 객체가 사용한 메모리 공간 자동관리
> 개발자가 메모리 관리를 직접할 필요 없음
> 변수 제거는 del()를 통해 제거 가능함
```

```python
* 비트 연산자
& 양변의 비트값 모두 1이 ㄹ경우에만 1을 반환
| 양변의 값 모두 0일 경우에만 0을 반환
^ 양변의 값이 다를 경우 1, 같을 경우 0 을 반환
~ 비트 값이 1일 경우 0, 0일경우 1를 반환
<< 좌변 값을 우변의 값 만큼 비트를 왼쪽으로 이동 # x=4 x<<1 #8
>> 좌변의 값을 우변의 값 만큼 비트를 오른쪽으로 이동 # x=8, x>>1 #4



a,b =2 -,3  # 0010, 0011
print(a&b) # 0010 & 0011 => 0010 #2
print(a|b)  # 0010 || 0011 => 0011 #3
print(a^b) #0010 ^ 0011 => 0001 #1
print(~a)   #~ 0010 => 1101 # -3 # 2의 보수를 구하는 연산

a=8 # 0000 1000
print(a>>1) # 0000 1000 >>1 => 0000 0100 #4
print(a<<1) # 0000 1000 <<1 => 0001 0000 #16

*주의 사항
괄호가 최우선
산술 연산이 비트 연산보다 우선
관계 연산이 논리 연산보다 우선
```
